```python
import matplotlib.pyplot as plt
```

```python
# Function and it's derivatives
def f(x):
    return (x + 3) ** 2

def df(x):
    return 2 * (x + 3)
```

```python
# Initial value of x 
x_old = 2
learning_rate = 0.1
precision = 0.0001
max_iters = 1000
```

```python
# To store progress
x_values = []
y_values = []
```

```python
for i in range(max_iters):
    grad = df(x_old)
    x_new = x_old - learning_rate * grad

    x_values.append(x_new)
    y_values.append(f(x_new))

    # Stop if changes is smaller than precision
    if abs(x_new - x_old) < precision:
        break

    x_old = x_new
```

```python
# Result
print(f"Local minima occurs at x = {x_new:.4f}")
print(f"Minimum value of function = {f(x_new):.4f}")
```

```python
# Plot
plt.plot(x_values, y_values, 'o-', label='Gradient Descent Path')
plt.title("Gradient Descent to Find Local Minima")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
