```python
import matplotlib.pyplot as plt
```

```python
x_old = 2
precision = 0.0001
lerning_rate = 0.1
iteration = 1000
```

```python
def f(x):
    return (x + 3) ** 2

def df(x):
    return 2 * (x + 3)

xvalues = []
yvalues = []

for i in range(iteration):
    grad = df(x_old)
    x_new = x_old - lerning_rate * grad
    xvalues.append(x_new)
    yvalues.append(f(x_new))

    if(abs(x_new - x_old)) < precision:
        break
    x_old = x_new

```

```python
print(f"Local minima occurs at x = {x_new:.4f}")
print(f"Minimum value of function = {f(x_new):.4f}")
```

```python
plt.plot(xvalues, yvalues, 'o--', label="Gradient Descent Algo")
plt.show()
```
